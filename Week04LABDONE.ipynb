3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
#NAME:SAHIL PANCHAL
#MISIS: M01097573
def passwordtoascii(password: str):
    astring=''
    for char in password:
        avalue = ord(char)
        astring += str(avalue)     #appending ASCII value of the char to the final string
    return astring

password = input("Enter your password: ")
print("Your ASCII-encoded password is:", passwordtoascii(password))
````

10. Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
#NAME:SAHIL PANCHAL
#MISIS: M01097573
def count_char(s):
    lower_case = 0
    upper_case = 0
    digits = 0
    special_symbols = 0
    #loop for fetching each character and appending its count accordingly 
    for char in s:
        if char.islower():
            lower_case += 1
        elif char.isupper():
            upper_case += 1
        elif char.isdigit():
            digits += 1
        else:
            special_symbols += 1

    print("Lowercase letters:", lower_case)
    print("Uppercase letters:", upper_case)
    print("Digits:", digits)
    print("Special symbols:", special_symbols)

# Example usage
example_password = "P@#yn26at^&i5ve"
count_char(example_password)
````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
#NAME:SAHIL PANCHAL
#MISIS: M01097573
import random
special_chars = "@#$%^&*!".strip()
def secret_word(name, surname):
    #slicing the first name and last name according to the format
    str1 = name[:3]
    str2 = surname[-2:]
    # Choose a random special character
    special = random.choice(special_chars)
    # Combine all parts
    secret_wrd = str1 + str2 + special
    return secret_wrd

def main():
    name = input("Enter your first name: ")
    surname = input("Enter your surname: ")

    # Generate and display the secret word
    secret = secret_word(name, surname)
    print("Your secret word is:", secret)
main()
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
#NAME:SAHIL PANCHAL
#MISIS: M01097573
def binaryToDecimal(binaryString):
    decimal_value = 0
    length = len(binaryString)
    
    for i in range(length):
        bit = binaryString[i]
        if bit == '1':
            # Calculate the power of 2 based on position
            power = length - 1 - i
            decimal_value += 2 ** power
        elif bit != '0':
            # Handling characters other thatn 0 and 1
            raise ValueError("Invalid binary digit found: " + bit)
    
    return decimal_value

binary_input = input("Enter a binary string: ")
try:
    result = binaryToDecimal(binary_input)
    print("Decimal value:", result)
except ValueError as e:
    print("Error:", e)
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
#NAME:SAHIL PANCHAL
#MISIS: M01097573
def binaryToHex(binaryValue):
    try:
        # Convert binary string to decimal integer
        decimal= int(binaryValue, 2)
        # Convert decimal integer to hexadecimal string
        hex_val = hex(decimal)[2:].upper()  # Remove '0x' prefix and capitalize
        return hex_val
    except ValueError:
        return "Invalid binary input"
binary_input = input("Enter a binary number: ")
hex_result = binaryToHex(binary_input)
print("Hexadecimal value:", hex_result)
````

17. (Decimal to binary) Write a function that converts a decimal number into a binary number as a string. Use the function header: def decimalToBinary(value):
Write a test program that prompts the user to enter a decimal integer value and displays the corresponding binary value.

````python
#NAME:SAHIL PANCHAL
#MISIS: M01097573
def decimalToBinary(value):
    # Converting decimal to binary using built-in bin() and stripping the '0b' prefix
    binary_string = bin(value)[2:]
    return binary_string

try:
    decimal_input = int(input("Enter a decimal integer: "))
    binary_result = decimalToBinary(decimal_input)
    print("Binary value:", binary_result)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
````
