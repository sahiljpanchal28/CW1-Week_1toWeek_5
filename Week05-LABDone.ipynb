### Activity 20: Managing Event Tickets

You are developing a system to manage event tickets for concerts or sports events. Each event has a limited number of tickets available, and users can purchase tickets if they are still available. You need to track which users have purchased tickets for specific events and how many are still available. What data structure would you use for this task, and why?

**Example data:**
- Available tickets: `{"Concert": 100, "Football Game": 50, "Theater Show": 75}`
- Purchasers: `{"Concert": ["John", "Emily"], "Football Game": ["Alex"], "Theater Show": ["Sam", "John"]}`

**Steps:**
1. Choose a data structure to track ticket availability and purchasers efficiently.
2. Write a program that handles ticket purchases and updates the list of purchasers.
3. Implement functions to check availability and view who has purchased tickets.
<hr>

`````python
#NAME: SAHIL JASMINKUMAR PANCHAL
#MISIS: M01097573
#BATCH: MONDAY
'''
DICTIONARIES ARE PREFFERED FOR THIS TASK BECAUSE OF THE FOLLOWING REASONS:
-  Fast lookup: Dictionaries allow constant-time access to ticket counts and purchaser lists.
-  Clear organization: Each event can store both ticket availability and purchaser data.
-  Easy updates: You can modify ticket counts and purchaser lists directly.
'''
events = {
    "Concert": {"available": 100, "purchasers": []},
    "Football Game": {"available": 50, "purchasers": []},
    "Theater Show": {"available": 75, "purchasers": []}
}

# Function to purchase a ticket
def purchase_ticket(event_name, user_name):
    if event_name in events:
        if events[event_name]["available"] > 0:
            events[event_name]["available"] -= 1
            events[event_name]["purchasers"].append(user_name)
            print(f"{user_name} successfully purchased a ticket for {event_name}.")
        else:
            print(f"Sorry, {event_name} is sold out.")
    else:
        print(f"Event '{event_name}' does not exist.")

# Function to check availability
def check_availability(event_name):
    if event_name in events:
        print(f"{event_name} has {events[event_name]['available']} ticket(s) left.")
    else:
        print(f"Event '{event_name}' does not exist.")

# Function to view purchasers
def view_purchasers(event_name):
    if event_name in events:
        print(f"Purchasers for {event_name}: {events[event_name]['purchasers']}")
    else:
        print(f"Event '{event_name}' does not exist.")
purchase_ticket("Concert", "John")
purchase_ticket("Concert", "Emily")
purchase_ticket("Football Game", "Alex")
purchase_ticket("Theater Show", "Sam")
purchase_ticket("Theater Show", "John")

check_availability("Concert")
view_purchasers("Theater Show")
`````

### Activity 21: Personal Finance App

You are creating a personal finance app that allows users to categorize their expenses and track spending by category (e.g., food, entertainment, travel). Users can add expenses, remove expenses, and see a summary of their spending. What data structure would you choose to store and organize the expenses, and why?

**Example data:**

```python
{"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}
```
Steps:

1. Choose a data structure that lets users easily add and track expenses by category.
2. Write a program that allows users to input expenses, remove them, and view their total spending by category.
3. Implement a function to display a summary of all spending.
<hr>

`````python
#NAME: SAHIL JASMINKUMAR PANCHAL
#MISIS: M01097573
#BATCH: MONDAY
expenses = {
    "Food": [25, 40, 15],
    "Entertainment": [50, 20],
    "Travel": [300, 150]
}

# Add an expense to a category
def add_expense(category, amount):
    if category in expenses:
        expenses[category].append(amount)
    else:
        expenses[category] = [amount]
    print(f"Added {amount} to {category}.")

# Remove an expense from a category
def remove_expense(category, amount):
    if category in expenses and amount in expenses[category]:
        expenses[category].remove(amount)
        print(f"Removed {amount} from {category}.")
    else:
        print(f"{amount} not found in {category}.")

# Display summary of spending
def display_summary():
    print()
    print("Spending Summary:")
    total = 0
    for category, amounts in expenses.items():
        category_total = sum(amounts)
        total += category_total
        print(f"{category}: AED {category_total}")
    print(f"Total Spending: AED {total}")

# Example usage
add_expense("Food", 30)
remove_expense("Travel", 150)
display_summary()
`````
